# Monte Carlo Simulation of Particle Transport and Decay

In this notebook, we demonstrate the core physics and capabilities of our particle simulation framework. 
We will utilize the high-performance vectorized `MonteCarloEngine` to model large populations.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Ensure the src module is in the path
sys.path.append(os.path.abspath('..'))
from src.simulation import MonteCarloEngine
from src.utils import plot_trajectories, plot_survival_curve, plot_msd, get_summary_statistics

import warnings
warnings.filterwarnings('ignore')
```

### 1. Initializing the Engine

We set up an engine to run a 2D diffusion process with a boundary.

```python
engine = MonteCarloEngine(
    num_particles=5000,
    dimensions=2,
    step_size=2.0,       # Diffusion character
    decay_constant=0.2,  # lambda
    dt=0.1,
    boundary_size=30.0,
    boundary_type='none', # Change to 'reflective' or 'absorbing' to see effects
    seed=42
)

results = engine.run(total_time=15.0)
stats = get_summary_statistics(results)
print(f"Survival: {stats['final_alive_percentage']:.2f}%")
print(f"Final MSD: {stats['final_msd']:.2f}")
```

### 2. Result Visualization

```python
fig_traj = plot_trajectories(results, num_to_plot=15, dimensions=2)
plt.show()
```

```python
fig_decay = plot_survival_curve(results, engine.decay_constant)
plt.show()
```

```python
fig_msd = plot_msd(results, engine.step_size, engine.dimensions)
plt.show()
```
